```python
#Dependencies
import os 
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Read data files
crime_2019 = "data/CrimeData-2019.csv"
crime_2020 = "data/CrimeData-2020.csv"
crime_2021 = "data/CrimeData-2021.csv"
crime_2022 = "data/CrimeData-2022.csv"
crime_2023 = "data/CrimeData-2023.csv"

# Read the crime data of each year
crime_2019 = pd.read_csv(crime_2019)
crime_2020 = pd.read_csv(crime_2020)
crime_2021 = pd.read_csv(crime_2021)
crime_2022 = pd.read_csv(crime_2022)
crime_2023 = pd.read_csv(crime_2023)

crime_2019
```

```python
# Combine the dataframes using concat
combined_df = pd.concat([crime_2019, crime_2020, crime_2021, crime_2022, crime_2023], ignore_index=True)
```

```python
combined_df
```

```python
#consistency counting -- looking for missing values
combined_df.info()
```

```python
combined_df['Neighborhood'].value_counts()
```

```python
combined_df['OffenseCategory'].value_counts()
```

```python
combined_df['OffenseType'].value_counts()
```

```python
missing = pd.DataFrame([combined_df.isnull().sum(), 100 * combined_df.isnull().mean()]).T
missing.columns=['count','%']
missing.sort_values(by='count')
```

```python
#dropping null rows
combined_df = combined_df.dropna(how ='any')
```

```python
#Checking consistency in dataframe after removing NAN on Neighborhood
combined_df.info()
```

```python
combined_df
```

```python
# this is to replace all the - with an empty space on the case number column
combined_df['CaseNumber'] = combined_df['CaseNumber'].apply(lambda x: x.replace('-',''))
```

```python
combined_df
```

```python
# drop duplicate in Case Number
combined_df = combined_df.drop_duplicates(['CaseNumber'], keep=False)
```

```python
combined_df.info()
```

```python
combined_df['Neighborhood'].unique()
```

```python
# Convert the 'OccurDate' column to datetime format
combined_df['OccurDate'] = pd.to_datetime(combined_df['OccurDate'])

# Extract the year and create a new column 'Year'
combined_df['Year'] = combined_df['OccurDate'].dt.year
```

```python
# save and export the edited .csv file 
combined_df.to_csv('updated-portland-oregon-crime-2019 till 2023.csv', index = False)
```

# Generating Exploratory Analysis & Other Summary Statistics

```python
# Read the CSV file
combined_crime1 = "updated-Datafile-Crime.csv"
# Load the CSV file into a DataFrame
combined_crime1 = pd.read_csv(combined_crime1)

```

```python

# Group the data by year and offense category and count occurrences
grouped_data = combined_crime1.groupby(['Year', 'OffenseCategory']).size().unstack(fill_value=0)

# Create a stacked bar plot
plt.figure(figsize=(20, 15))
ax = grouped_data.plot(kind='bar', stacked=True)

plt.title('Crime Offense Categories over the last Five Years')
plt.xlabel('Year')
plt.ylabel('Number of Occurrences')

# Move the legend away from the main graph using bbox_to_anchor
legend = ax.legend(title='Offense Category', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()

plt.show()
```

```python
# Group the data by Neighborhood and offense category and count occurrences
grouped_data = combined_crime1.groupby(['Neighborhood', 'OffenseCategory']).size().unstack(fill_value=0)

grouped_data
```

```python
# Group the data by Crime Against and offense category and count occurrences
grouped_data = combined_crime1.groupby(['CrimeAgainst', 'OffenseCategory']).size().unstack(fill_value=0)

# Create a stacked bar plot
plt.figure(figsize=(20, 15))
ax = grouped_data.plot(kind='bar', stacked=True)

plt.title('Crime Offense Categories over the last Five Years')
plt.xlabel('Year')
plt.ylabel('Number of Occurrences')

# Move the legend away from the main graph using bbox_to_anchor
legend = ax.legend(title='Offense Category', loc='upper left', bbox_to_anchor=(1, 1))

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
# Read the CSV file
df2 = "updated-Datafile-Crime.csv"
# Load the CSV file into a DataFrame
df2 = pd.read_csv(df2)

# Convert the numbers to formatted strings with colons
df2['OccurTime'] = df2['OccurTime'].apply(lambda x: '{:02d}:{:02d}:00'.format(int(x // 100), int(x % 100)))

# Now the 'TimeData' column contains formatted time strings with colons
print(df2)
```

```python
# Convert the 'OccurTime' column to datetime format
df2['OccurTime'] = pd.to_datetime(df2['OccurTime'], format='%H:%M:%S')

# Convert the datetime format back to string format with the same '%H:%M:%S' format
df2['NewOccurTimeFormat'] = df2['OccurTime'].dt.strftime('%H:%M:%S')

# Now the 'NewOccurTimeFormat' column contains the time data in the desired format 'H:M:S'
print(df2)
```

```python
# Save the updated DataFrame back to a CSV file
df2.to_csv('updated-Datafile-Crime2.csv', index=False)
```

```python
# Read the CSV file
df3 = "Updated-Datafile-Crime2.csv"
# Load the CSV file into a DataFrame
df3 = pd.read_csv(df3)
```

```python
df3['OccurTime'] = df3['OccurTime'].str.replace(' ', '')
```

```python
# Define function to categorize time periods
def categorize_time_period(time_str):
    hour = int(time_str.split(':')[0])  # Extract the hour from the time string
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Create a new column 'TimePeriod' based on the categorized time periods
df3['TimePeriod'] = df3['NewOccurTimeFormat'].apply(categorize_time_period)

df3.head()
```

```python
#Drop columns
columns_to_drop = ['Address','OccurTime', 'OpenDataX','OpenDataY', 'ReportDate']  
df3.drop(columns=columns_to_drop, axis=1, inplace=True)

# Rename a column
#df3.rename(columns={'NewOccurTimeFormat': 'OccurTime'}, inplace=True)


df3.head()
```
